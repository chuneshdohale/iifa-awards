```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content,"html5lib")
soup
```

```python
data=soup.find_all(class_="wikitable sortable")
len(data)
```

```python
rich_men=data[0]
rich_men
```

```python
rich_women=data[1]
rich_women
```

```python
heads=[d.get_text().strip() for d in rich_men.find_all("th")]
heads
```

```python
table_data=rich_men.find_all("tr")[1:]
table_data
```

```python
p=[[d.get_text().strip() for d in f.find_all("td")] for f in table_data]
p
```

```python
l=rich_men.find_all('span')
U=[i.find('img')['alt'] for i in l ][::2]
len(U)
```

```python
for d in range(len(p)-1):
    if p[d][2]=='':
        p[d][2]=U[d]

p[-1][2]=U[-1]
p
```

```python
d={K:[] for K in heads}

d 
```

```python
for i in p:
    d['Rank'].append(i[0])
    d['Name'].append(i[1])
    d['Wealth\nChange'].append(i[2])
    d['Net worth (USD)'].append(i[3])
    d['Company'].append(i[4])
    d['Sources of wealth'].append(i[5])
d
```

```python
df=pd.DataFrame(d)
df
```

```python
df.to_csv("Top_25_richest_men.csv",index=False)
```
